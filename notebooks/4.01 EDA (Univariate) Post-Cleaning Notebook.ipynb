### Games Started Distribution
Looking at the distribution of games started, it seems fairly uniform, with a slight decrease after 12 games played.
```python
eda_viz_post.gs_distribution(savefig = True)
```

### Targets Distribution
The distribution of targets is fairly close to normal, which should lend itself well to a linear regression model, and therefore can be left as is.

```python
eda_viz_post.tgt_distribution(savefig = True)
```

### WR Target Share Distribution
The distribution of target share, on the other hand, has some larger outliers on the right side (Zay Jones and Dwayne Bowe). This large right tail indicates that we should most likely apply some kind of function/filter to said feature, and when looking at both the boxcox square root function, as well as a normal square root function, both seem to give more normally shaped distributions. These transformations will be applied with a function transformer in our modeling pipeline.

```python
eda_viz_post.tgt_share_distribution(savefig = True)
```

```python
df_model[df_model['WR Tgt Share'] > .5]
```

```python
eda_viz_post.sqrt_tgt_share_distribution(savefig = True)
```
